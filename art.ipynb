```python
import matplotlib.pyplot as plt
import os
import numpy as np  
%matplotlib inline
import math
import random
```

```python
points_num=150
```

```python
points_arr=[]
for x in range(int(-1*points_num/2), int(points_num/2)):
    for y in range(int(-1*points_num/2), int(points_num/2)):
        points_arr.append((x, y))
```

```python
X, Y = zip(*points_arr)
X=np.array(X)
Y=np.array(Y)
#divide X and Y by 75
X=X/75
Y=Y/75
# Create a scatter plot
plt.scatter(X, Y, s=1)

# Set the title and labels
plt.title('Plot of Points')
plt.xlabel('X')
plt.ylabel('Y')

# Show the plot
plt.show()
```
