```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import to_rgb

def hex_to_rgb(hex_color):
    return [int(hex_color[i:i+2], 16) for i in (1, 3, 5)]

def rgb_to_hex(rgb_color):
    return "#" + "".join([f"{int(i):02x}" for i in rgb_color])

def blend_colors(t, color1, color2):
    rgb1 = np.array(hex_to_rgb(color1))
    rgb2 = np.array(hex_to_rgb(color2))
    blended = ((1 - t) * rgb1 + t * rgb2).astype(int)
    return rgb_to_hex(blended)

color1 = "#000000"
color2 = "#ffffff"

x = np.linspace(0, 1, 100)
y = np.linspace(0, 1, 100)
grid_x, grid_y = np.meshgrid(x, y)

color = np.array([[blend_colors((np.sin((i+j)/2)+np.cos((i+j)/2))/2, color1, color2) for i, j in zip(row_x, row_y)] for row_x, row_y in zip(grid_x, grid_y)])

#plot color
fig, ax = plt.subplots()
ax.imshow([[to_rgb(c) for c in row] for row in color], extent=(-1, 1, -1, 1), origin='lower')
plt.show()
```
